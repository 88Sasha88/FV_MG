```python
def FindSelect(omega, waves):
    errorLoc = 'ERROR:\nOperatorTools:\nFindNullspace:\n'
    levels = omega.levels
    alias = omega.alias
    nh_max = omega.nh_max
    degFreed = omega.degFreed# [::-1][0]
    fixWaves = np.zeros((nh_max, degFreed), float)
    nh_min = omega.nh_min
    nhs = omega.nh
    nh_mid = nhs[-2]
    hs = omega.h
#     print(nh_mid)
#     print(degFreed)
#     mode = nh_mid - 1

    
    
    # num = degFreed - mode - 1
    addOn = degFreed - nh_mid - 1# - 2
    
    iters = int((nh_max / 2) + 1)
    # mode is the index 
    # addOn = int(addOn * np.heaviside(addOn, 0))
    
    np.fill_diagonal(fixWaves[:, :-addOn], 1) # This won't work for low degFreed.
#     print(mode, fixWaves)
    # Iterate through number of wavenumbers left
    addOn = addOn
    mode = nh_mid + 1
    wavenum = int(mode / 2)
    print(wavenum, iters)
    print('START')
    j = mode# wavenum
    span = min(nh_mid + 2, degFreed)
    span0 = span
    print('addOn:', addOn)
    for i in range(wavenum + 1, iters):
        fixWavesTemp = fixWaves + 0
        span = span + 1
        print('spans outer:', span, span0)
        print('interation:', i)
        print('addOn Before:', addOn)
        # If any more wavenumbers are needed, otherwise, stop the outermost for loop.
        if (addOn > 0):
            print('A')
            if (addOn % 2 == 1):
                print('B')
                fixWavesTemp[-1, -1] = 1
                addOn = addOn - 1
                # mode = mode + 1
            else:
                print('C')
                cond = True
                span = span + 1
                while (cond):#(span != span0):
                    fixWavesTemp = fixWaves + 0
                    fixWavesTemp[mode, j] = 1
                    fixWavesTemp[mode + 1, j + 1] = 1
                    print(mode, j)
                    print(mode + 1, j + 1)
                    print(fixWavesTemp)
                    wavesTemp = waves @ fixWavesTemp
                    WTW = wavesTemp.transpose() @ wavesTemp
                    eigvals, eigvecs = LA.eig(WTW)
                    eigvals = abs(np.round(eigvals, 12))
                    span0 = degFreed - len(np.where(eigvals == 0)[0])
                    cond = any(eigvals == 0)
                    print(eigvals)
                    print('spans inner:', span, span0)
                    print('')
                    # ADD AN ERROR CATCH HERE FOR IN CASE YOU CAN'T MANAGE TO SPAN THE SPACE!!!
                    mode = mode + 2
                j = j + 2
                addOn = addOn - 2
            fixWaves = fixWavesTemp
            
        else:
            break
        print(fixWaves)
        print('')
        print('addOn After:', addOn)
#     if (degFreed < nh[])
    print('END')
    print('')
    print(fixWaves)
    return fixWaves
```
